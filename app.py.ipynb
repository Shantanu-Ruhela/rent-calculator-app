```python
## input we need from the user
# Total rent
# Total food ordered for snacking
# Electricity units spend
# charge per unit
# Persons living in room/flat

## Output
# Total amount you've to pay is

rent = int(input("Enter your hostel/flat rent = "))
food = int(input("Enter the amount of food ordered = "))
electricity_spend = int(input("Enter the total of electricity spend = "))
charge_per_unit = int(input("Enter the charge per unit = "))
persons = int(input("Enter the number of persons living in room/flat = "))

total_bill = electricity_spend * charge_per_unit

output = (rent + food + total_bill) // persons
print(output)
```

```python
import streamlit as st
```

```python
st.title("Rent Calculator App")

# inputs
rent = st.number_input("Enter your hostel/flat rent",min_value=0)
food = st.number_input("Enter the amount of food ordered",min_value=0)
electricity_spend = st.number_input("Enter electricity units spent",min_value=0)
charge_per_unit = st.number_input("Enter per unit charge",min_value=0)

persons = st.number_input("Enter number of persons",min_value=0)

# Calculation
if st.button("calculation"):
    total_bill = electricity_spend * charge_per_unit
    total = (rent + food + total_bill) // persons
    st.success(f"Each person should pay :Rs{total}")
```
